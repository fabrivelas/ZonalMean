```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import ScalarFormatter, FuncFormatter, AutoMinorLocator

def log_to_real_pressure(plev_log):
  """Converts pressure from logarithmic scale to real scale.

  Args:
    plev_log: Pressure in logarithmic scale.

  Returns:
    Pressure in real scale.
  """

  return np.power(10, plev_log)
```

```python
data = pd.read_csv('nhant.lsn', header=6,)
```

```python
ncdata = xr.open_dataset('cira.nc')
```

```python
meanTprof = ncdata.mean('time').mean('latitude')
# Calculate altitudes from pressure and temperature
R_dry = 287.05  # Specific gas constant for dry air in J/(kgÂ·K)
T0 = 273.15  # Reference temperature in K
P0 = 1000.0  # Reference pressure in hPa

pressure = meanTprof.plev
temperature = meanTprof.ta
altitudes = ((-R_dry * T0) / 9.8) * (1 - (pressure / P0) ** (R_dry * 9.8 / (temperature + T0)))

print(meanTprof.zg/1000)
print(altitudes)
```

```python
# Create a log-scaled y-axis by specifying the scale attribute
fig, ax = plt.subplots()
meanTprof.ta.plot(ax=ax, yscale='log', y='plev', visible=False)

# Set labels and title
plt.xlabel('Temperature (K)')
plt.ylabel('approximate Pressure (hPa)')
#plt.title('Temperature vs Pressure Level (Log-scaled)')
# plt.gca().invert_yaxis()
# Reverse the y-axis
ax.invert_yaxis()

# Format the y-axis tick labels without exponents
#ax.yaxis.set_major_formatter(ScalarFormatter())
# Create a custom format for the y-axis tick labels
# Create a custom format for the y-axis tick labels
def custom_format(value, pos):
    if value < 1:
        return f"{value:.4g}"
    else:
        return f"{int(value)}"


# Add a second y-axis for altitudes on the right
ax2 = ax.twinx()
ax2.set_ylabel('Altitude (km)')
ax2.plot(meanTprof.ta, meanTprof.zg/1000, 'r', label='Altitude')

# Add minor ticks to y-axes
ax2.xaxis.set_minor_locator(AutoMinorLocator())
#ax.xaxis.set_minor_locator(AutoMinorLocator())

ax2.axhline(y=9, color='b', linestyle='--', label='9 km Altitude')

# Create a secondary x-axis at the top to match temperature
ax_top = ax.secondary_xaxis('top')

ax3 = ax.twiny()
ax3.plot(meanTprof.ta, meanTprof.zg/1000, 'r', label='Altitude', visible='false')
ax3.xaxis.set_minor_locator(AutoMinorLocator())
ax2.yaxis.set_minor_locator(AutoMinorLocator())
```
