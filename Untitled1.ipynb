```python
import xarray as xr
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

ncdata = xr.open_dataset('cira.nc')

mean_prof = ncdata.mean('time').mean('latitude')

# Calculate altitudes from pressure and temperature
#R_dry = 287.05  # Specific gas constant for dry air in J/(kgK)
#T0 = 273.15  # Reference temperature in K
#P0 = 1000.0  # Reference pressure in hPa

#pressure = meanTprof.plev
#temperature = meanTprof.ta
#altitudes = ((-R_dry * T0) / 9.8) * (1 - (pressure / P0) ** (R_dry * 9.8 / (temperature + T0)))

#print(meanTprof.zg/1000)
#print(altitudes)

# Create a log-scaled y-axis by specifying the scale attribute
fig, ax = plt.subplots()
mean_prof.ta.plot(ax=ax, yscale='log', y='plev', visible=False)

# Set labels and title
plt.xlabel('Temperature (K)')
plt.ylabel('Approximate Pressure (hPa)')
# Reverse the y-axis
ax.invert_yaxis()

# Add a second y-axis for altitudes on the right
ax2 = ax.twinx()
ax2.set_ylabel('Altitude (km)')
ax2.plot(mean_prof.ta, mean_prof.zg/1000, 'r', label='T')

# Add minor ticks to y-axes
ax2.yaxis.set_minor_locator(AutoMinorLocator())

# Add horizontal lines at specific altitudes with labels
pausen = ['polar Tropopause', 'Tropopause', 'Stratopause', 'Mesopause']
altitudes = [9, 15, 47, 92]
for i, altitude in enumerate(altitudes):
    ax2.axhline(y=altitude, color='b', linestyle='--', label=pausen[i])
    ax2.text(plt.xlim()[1], altitude, pausen[i], va='bottom', ha='right')



# Create a secondary x-axis at the top to match temperature
ax_top = ax.secondary_xaxis('top')

# Add minor ticks to x-axes (temperature axis)
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax2.xaxis.set_minor_locator(AutoMinorLocator())
ax_top.xaxis.set_minor_locator(AutoMinorLocator())


# Apply tight layout
plt.tight_layout()
ax.set_ylim(bottom=1.013000e+03)
ax2.set_ylim(bottom=0)


# Show the plot
plt.show()
```
